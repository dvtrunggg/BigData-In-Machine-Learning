```
from pyspark import SparkContext
sc =SparkContext(master="local", appName="New Spark Context") 
```

```
data = sc.textFile( "t8.shakespeare.txt").cache() 
```

```
numAs = data.filter(lambda s: "face" in s).count()
numBs = data.filter(lambda s: "world" in s).count()

print("Lines with 'face': %i, lines with 'world': %i" % (numAs, numBs))
```

**Cho biết tập tin có bao nhiêu line?**

```
print("Lines: %i" % (len(data.collect())))
```

**Cho biết word "love" xuất hiện trên bao nhiêu line?**

```
numLoves = data.filter(lambda s: 'love' in s).count()
print("Lines with 'love': %i" % (numLoves))
```


**Cho biết trong tập tin có xuất hiện số (bất kỳ) không?\
Nếu có thì xuất hiện trên bao nhiêu line? \
Liệt kê các line đó?**

```
import re
pattern = "[\s|\S]*\d{1,}[\s|\S]*"
```

```
numbers = data.filter(lambda s: re.match(pattern,s)).count() 
numbers
```

```
if numbers > 0 :
 print("Lines with numbers: %i" % (numbers))
 number_lines = data.filter(lambda s: re.match(pattern,s))
 i = 0
 for line in number_lines.collect():
  print(i,":",line)
  i+=1
else:
 print("No number in this text.")
```
